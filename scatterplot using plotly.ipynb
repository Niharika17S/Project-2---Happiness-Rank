```python
import pandas as pd
import numpy as np

import plotly as py
import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)

from ipywidgets import widgets
from IPython.display import display
from plotly.widgets import GraphWidget
```

```python
data2015_df = pd.read_csv("Data/2015.csv")
```

```python
data2015_df.head()
```

```python
# Using .rename(columns={}) in order to rename columns
renamed2015_df = data2015_df.rename(columns={"Happiness Score":"Happiness_Score","Economy (GDP per Capita)":"Economy_GDP","Health (Life Expectancy)":"Health" })

renamed2015_df.head()
```

```python
trace1 = go.Scatter(
                    x=renamed2015_df['Economy_GDP'], y=renamed2015_df['Happiness_Score'], 
                    mode = 'markers',
                    #name='plot' # Additional options
                   )
trace2 = go.Scatter(
                    x=renamed2015_df['Health'], y=renamed2015_df['Happiness_Score'], 
                    mode = 'markers',
                    marker=dict(
                    size=16,
                    color = np.random.randn(600), #set color equal to a variable
                    colorscale='Viridis',
                    showscale=True
                    ),
                    name='Health'
                    )
data = [trace1,trace2]
layout = go.Layout(title='GDP vs Happiness',
                   xaxis = dict(title = "Economy(GDP)"),
                   yaxis = dict(title = "Happiness Score"),
                   plot_bgcolor='rgb(230, 230,230)')
```

```python
w = widgets.Dropdown(
    options= list(renamed2015_df['Country'].unique()),
    value='Iceland',
    description='Country',
)
```
